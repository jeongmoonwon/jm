```python
t0.shape 
```

```python
t1 = torch.sum(t, dim=1, keepdim=True) 
t1
```

```python
# eq: 각 원소가 같은지 다른지 비교하기
```

```python
torch.eq(torch.tensor([[1, 2], [3, 4]]), torch.tensor([[1, 1], [4, 4]]))
```

```python
# 텐서t와 새로운 텐서를 비교
t.eq(torch.tensor([[1, 2], [3, 4]])) 
```

```python
# True를 1로, False를 0으로 count해서 더함
t.eq(torch.tensor([[1, 2], [3, 4]])).sum() 
```

```python
# ----- 두 개의 텐서 합치기 -----
```

```python
t1 = torch.tensor([[1, 2],
                   [3, 4]])
t2 = torch.tensor([[5, 6],
                   [7, 8]])
```

```python
torch.cat((t1, t2), dim=0) # dim=0인 경우
```

```python
torch.cat((t1, t2), dim=1) # dim=1인 경우
```

```python
torch.stack((t1, t2)) # dimension을 유지 하면서 붙여주는 것
```

```python
# ----- 텐서 차원 축소 -----
```

```python
t = torch.randn((2,3,4))
t
```

```python
# argmax : 가장 큰 값을 가지는 index를 뽑는 함수 & dim=1은 1st dim에서 비교하겠다는 의미
t_argmax1 = torch.argmax(t, dim=1) 
t_argmax1
```

```python
t_argmax1.shape
```

```python
t_argmax2 = torch.argmax(t, 1, keepdims=True)
```

```python
t_argmax2.shape
```
