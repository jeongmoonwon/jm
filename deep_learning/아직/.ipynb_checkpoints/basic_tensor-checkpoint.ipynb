## 텐서 만들기
tensor란 다차원 array로 쉽게 말해 행렬이라고 생각하면 된다!

```python
import torch
import numpy as np
```

- 임의의 텐서 만들기

```python
torch.Tensor(2,3) 
```

```python
torch.rand(2,3)
```

- element를 직접 지정해서 만들기

```python
torch.tensor([[1,2,3], [4,5,6]]) 
```

```python
torch.tensor(data=[1.0, 2.0], requires_grad=True) 
# float으로 입력해야 함
# requires_grad : 텐서에 대한 기울기를 저장할지 여부
```

- numpy array에서 tensor로 바꾸기

```python
n = np.array([[1,2],[3,4]])
type(n)
```

```python
t = torch.Tensor(n)
type(t)
```

- 0, 1로 채워진 텐서 만들기

```python
# 2x2로 된 영행렬 만들기
torch.zeros(2, 2) 
```

```python
# 2x2이고 모든 원소가 1인 행렬 만들기
torch.ones(2, 2) 
```

- 인자로 들어오는 텐서와 형태가 같으면서 0 혹은 1로 채워진 텐서 만들기

```python
t = torch.rand(4,5)
t0 = torch.zeros_like(t)
t1 = torch.ones_like(t)
```

```python
t0
```

```python
t1
```

## 텐서 변형하기

```python
t = torch.rand(12,1) 
t.size()
```

- reshape, permute

```python
t1 = t.reshape(3, 4, 1)
t2 = t1.permute(1, 2, 0) # 4, 1, 3
```

```python
t1
```

```python
t2
```

- squeeze, unsqueeze

```python
# 1인 차원을 제거
t.squeeze().size()
```

```python
# 특정 위치에 1인 차원을 추가. 이 경우 두번째 차원에 1을 추가
t.unsqueeze(1).size()
```

- view_as

```python
# size가 12인 텐서t를 size가 3,4,1인 텐서t1처럼 나타내기
t.data.view_as(t1) 
```

## 텐서 연산

```python
# 새로운 텐서 생성
t = torch.tensor([[1,2], [4,5]])
```

- sum
    - https://velog.io/@reversesky/torch.sum에-대해-알아보자

```python
# 텐서t의 모든 원소를 sum
torch.sum(t)
```

```python
# 특정 dim에 대하여 sum
t0 = torch.sum(t, dim=0, keepdim=True) # keepdim : 결과로 나오는 텐서가 차원을 그대로 유지하게 할 것인지
t0
```

```python
t0.shape 
```

```python
t1 = torch.sum(t, dim=1, keepdim=True) 
t1
```

```python
t1.shape 
```

- max

```python
t.max(1) # 괄호 안의 숫자는 dimension을 의미
```

```python
t.max(0) 
```

```python
# 최대값인 원소
t.max(0)[0] 
```

```python
# 최대값인 원소의 인덱스
t.max(0)[1] 
```

```python
t.max(0)[0].shape 
```

```python
# keepdim 여부에 따라 사이즈가 달라짐
t.max(0, keepdim=True)[0].shape 
```

- eq: 각 원소가 같은지 다른지 비교하기

```python
torch.eq(torch.tensor([[1, 2], [3, 4]]), torch.tensor([[1, 1], [4, 4]]))
```

```python
# 텐서t와 새로운 텐서를 비교
t.eq(torch.tensor([[1, 2], [3, 4]])) 
```

```python
# True를 1로, False를 0으로 count해서 더함
t.eq(torch.tensor([[1, 2], [3, 4]])).sum() 
```

- 그 외

```python
# 각 원소를 제곱
torch.pow(t, 2) 
```

```python
# exponential
torch.exp(t) 
```

```python
# sigmoid 계산
torch.sigmoid(t) # torch.special.expit(t)도 가능
```

```python
# matrix multiplication : [n, m] x [m,p] = [n,p] 를 구현
torch.mm(t, t) 
```

```python
# 벡터와의 연산
v = torch.ones((2,1)) # 만약 2x1 벡터가 아니라 3x1 벡터라면? 에러가 날 것이다!
t+v 
```

## 텐서 차원 축소

```python
t = torch.randn((2,3,4))
t
```

```python
# argmax : 가장 큰 값을 가지는 index를 뽑는 함수 & dim=1은 1st dim에서 비교하겠다는 의미
t_argmax1 = torch.argmax(t, dim=1) 
t_argmax1
```

```python
t_argmax1.shape
```

```python
t_argmax2 = torch.argmax(t, 1, keepdims=True)
```

```python
t_argmax2.shape
```
