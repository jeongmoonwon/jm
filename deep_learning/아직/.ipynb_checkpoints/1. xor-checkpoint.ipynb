```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# sigmoid 함수 정의
def sigmoid(x):
    return 1.0/(1.0 + np.exp(-x))

# parameter 초기값 지정
def init_network(input_dim, hidden_dim, output_dim):
    W1 = np.random.randn(input_dim, hidden_dim) # 2행 2열 (w11,w12,w21,w22)
    b1 = np.random.randn(hidden_dim) # bias 2개
    W2 = np.random.randn(hidden_dim, output_dim ) # 2행 1열 (v1,v2)
    b2 = np.random.randn(output_dim) # bias 1개
    return W, b, V, c

# 순전파
def forward(x, W, b, V, c):
    z = sigmoid(np.matmul(x, W) - b) # hidden activity
    y = sigmoid(np.matmul(z, V) - c) # 출력값 y를 predict
    return z, y

# 역전파
def backward(x, true_y, W, b, V, c, lr=0.01): # lr는 learning rate
    z, y = forward(x, W, b, V, c) # prediction구하기 (이걸 구해야 에러를 구할 수 있음)
    error_y = (true_y - y)*y*(1-y) # y*(1-y) = 시그마'(x)
    error_z = error_y.dot(V.T)*z*(1-z) # v라는 가중치를 가지고 redistribution, z*(1-z) = 시그마'(zj)
    
    V = V + lr * np.matmul(z.T, error_y)
    W = W + lr * np.matmul(x.T, error_z)
    c = c + lr * np.sum(error_y,axis=0)*(-1.)
    b = b + lr * np.sum(error_z,axis=0)*(-1.)
    return W, b, V, c 
```

```python
# XOR 문제 정의
x = np.array([[0,0], [0,1],[1,0],[1,1]])
true_y = np.array([[0.], [1.], [1.], [0.]])
epoch_num = 10000 # 업데이트 횟수

# 초기값
W, b, V, c = init_network(input_dim=2, hidden_dim=2, output_dim=1) 
# 정답
# W = np.array([[1.,1.],[1.,1.]])
# b = np.array([0.5, 1.5])
# V = np.array([[1.],[-1.]])
# c = np.array([0.5])
W
```

```python
epoch_dat = []
error_dat = [] # 각 epoch에 해당하는 에러
for epoch in range(epoch_num):
    W, b, V, c = backward(x, true_y, W, b, V, c, lr=0.1) # 업데이트
    z, y = forward(x, W, b, V, c) # y값 대시 예측
    error = true_y - y # 에러 구하기
    train_error = np.sum(error*error) # 모든 샘플에 대한 에러 더하기
    epoch_dat.append(epoch)
    error_dat.append(train_error)        
    
    if epoch % 1000 == 0: # 천 번에 한 번은 결과를 보여줘라
        print( 'epoch: %d/%d, error: %.3f'%(epoch, epoch_num, train_error))

plt.plot(epoch_dat, error_dat, '-*')
plt.show()
```
