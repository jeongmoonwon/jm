```python
# if절처럼 사용하기
score = [60, 75, 85, 91, 100]
grade = []

for i in score:
    condlist = [i >= 90,
                (i < 90) & (i >= 80),
                i < 80]
    choicelist = ['A','B','C']
    grade.append(np.select(condlist, choicelist))
    
grade
```

```python
# ----- Einsum -----
```

```python
a = np.array([[1,2,3], [4,5,6]])
a
```

```python
# transpose
np.einsum("ij->ji", a)
```

```python
# matrix sum to scalar
np.einsum("ij->", a)
```

```python
# matrix column or row sum
np.einsum("ij->i", a)
```

```python
np.einsum("ij->j", a)
```

```python
# diagonal, Trace
b = np.eye(4)
b
```

```python
np.einsum('ii->i', b)
```

```python
np.einsum('ii->', b)
```

```python
x = np.array([-1, -10, -100])
y = np.array([1, 10, 100])

# Dot Product
np.einsum('i,i->', x, y)
```

```python
# Outer product
np.einsum('i,j->ij', x,y)
```

```python
# element-wise product of vector
np.einsum('i,i->i', x, y)
```

```python
# element-wise product of matrix
a = np.arange(6).reshape((2, 3))
b = np.arange(6).reshape((2, 3))
a, b
```

```python
np.einsum('ij,ij->', a, b)
```

```python
# Matrix-Vector multiplication
a = np.array([[1,2,3], [4,5,6]])
b = np.array([-1, -10, -100])
np.einsum('ij,j->i', a, b)
```

```python
# Matrix-Matrix Multiplication
a = np.array([[1,2,3], [4,5,6]])
b = a.transpose()
np.einsum('ik,kj->ij', a, b)
```

```python
# Batched Matrix Multiplication
a = np.random.random(size=(3, 4, 3))
b = np.random.random(size=(3, 3, 2))
np.einsum('bik,bkj->bij',a, b)
```
