```python
# ----- pytz ----- 
# 세계 시간대 정의
```

```python
import pytz
from datetime import datetime
import time

time_zone = pytz.timezone('Asia/Seoul')
current_time = datetime.now(time_zone)
current_time_seconds = int(time.mktime(current_time.timetuple()))
seconds_to_timestring = datetime.fromtimestamp(current_time_seconds).strftime('%Y-%m-%d %H:%M:%S')

print(time_zone)
print(current_time_seconds)
print(seconds_to_timestring)
```

```python
import logging

log_output_path = '/Users/jeongmoonwon/Desktop/jm/python_code'

# 로그를 저장할 형식
# level=logging.INFO : INFO 레벨 이상은 출력 하도록 설정 (DEBUG 출력 안함)
logging.basicConfig(filename=log_output_path,
                    level=logging.INFO, 
                    format='[%(levelname)s]%(asctime)s| %(message)s', 
                    datefmt='%m-%d %H:%M:%S')
```
